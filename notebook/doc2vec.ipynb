```python
from functools import wraps
import time
def timer(func) :
    @wraps(func)
    def wrapper(*args, **kargs) :
        start = time.time()
        result = func(*args,**kargs)
        process_time =  time.time() - start
        print(f"elapsed time: {process_time} in {func.__name__}")
        return result
    return wrapper
```

```python
from nltk import tokenize
@timer
def nltk_word_toknize(docs):
    for _ in range(10):
        words = [
            [tokenize.word_tokenize(s) for s in tokenize.sent_tokenize(doctxt)]
            for doctxt in docs
        ]
nltk_word_toknize(docs)
```

```python
spacy_tokenizer = SpacyWordSplitter(pos_tags=True)

@timer
def spacy_word_toknize(docs):
    for _ in range(10):
        words = [
            [spacy_tokenizer.split_words(s) for s in tokenize.sent_tokenize(doctxt)]
            for doctxt in docs
        ]
spacy_word_toknize(docs)
```

```python
from itertools import chain
@timer
def tagdocs(docs):
    for k in range(10):
        tagged = [
            TaggedDocument(list(chain.from_iterable(doc)), [f"{k:03d}"])
            for doc in docs
        ]
    return tagged
tagged = tagdocs(docs)
```

```python
m = Doc2Vec(tagged, vector_size=128, window=3, min_count=1,
                             dm=1, hs=0, negative=10, dbow_words=1,
                             workers=4)
```
